```python
import everest
import matplotlib.pyplot as plt
import numpy as np
```

```python
star = everest.Everest(202073124)

#Source: https://ui.adsabs.harvard.edu/abs/2020svos.conf..437J/abstract
#202862302 period = 1.956 Cepheid
#217987553 period = 13.448
    
#Source: https://arxiv.org/pdf/1909.00446.pdf
#205969073 period = 0.509065 RRLyr 
#205905693 period = 0.486393 RRLyr-BL

# sources: https://www.aanda.org/articles/aa/full_html/2015/07/aa25889-15/aa25889-15.html
#201143426 period=8.334897 Sinosuid
#201156868 period = 2.266612 Sinosuid
#201160323 period = 22.299969 EB
#201161898 period = 2.642426 QP
```

```python
plt.plot(star.time, star.flux, '.')
# plt.yscale('log')
```

```python
star.cbv_num = 0
star.compute()
```

```python
y = star.fcor[(star.time<1899) | (star.time>1936)]
x = star.time[(star.time<1899) | (star.time>1936)]
# x = star.time#np.delete(star.time, star.mask)#
# y = star.fcor#np.delete(star.fcor, star.mask)#

p = 1.677689
phase = (x/p)- np.fix(x/p)

```

```python
plt.plot(star.time, star.fcor,'b.')
```

```python
plt.plot(star.time[(star.time<1900) | (star.time>1936)], star.fcor[(star.time<1900) | (star.time>1936)],'b.')
plt.xlim(1897,1900)
```

```python
fig, axs = plt.subplots(1,2,figsize=(25,10))


# axs[0].errorbar(np.delete(star.time, star.mask), np.delete(star.fcor, star.mask),yerr=np.delete(star.fraw_err, star.mask), fmt='o')
axs[0].plot(x, y,'b.')
# axs[0].set_ylim(0,8800)
axs[0].set_ylabel('Flux', size=25)
axs[0].set_xlabel('Time (days)', size=25)
axs[0].set_title('K2 light curves',size=30)
axs[1].plot(phase, y, 'b.',phase+1, y,'b.')
# axs[1].set_ylim(0,8800)
axs[1].set_xlabel('Phase', size=25)
axs[1].set_title('K2 phased light curves',size=30)


axs[0].tick_params(axis='both', which='major', labelsize=15)
axs[1].tick_params(axis='both', which='major', labelsize=15)
plt.tight_layout()

# plt.savefig('./../../../KELT_ML_Project/AI_expert_sessions/K2_11.png')
```
